# Aufgabe 2.1.1
1-D Furier - Basisfunktion , Orthogonalit√§t

Die 1-D Furier - Basisfunktion ist geggeben durch:

$$
\varphi_n(x) = e(iq_nx)
$$
wobei $ x \in [0, L]$ und $q_n = 2\pi n / L$.


Ist das Skalarproduktgleich von zwei Basisvektoren gleich Null, stehen diese Orthogonal zueinander.
Wir benutzen in unserem Fall $\varphi_n , \varphi_m$.

$$
(\varphi_n,\varphi_m) = \frac{1}{L} \int_0^L \varphi_n^*  \varphi_m dx
$$

$$(\varphi_n , \varphi_m )  = \frac{1}{L} \int_0^L e^{-i\frac{2 \pi n}{L}x} \cdot e^{i\frac{2 \pi m}{L}x} dx$$

$$(\varphi_n , \varphi_m ) = \frac{1}{L} \int_0^L e^{i\cdot(m-n)\cdot\frac{2 \pi x}{L}} dx$$


Sollte $m = n$ erhalten wird als L√∂sung die 1.

Sollte $m \neq n$ erhalten wir als L√∂sung die 0. Dies ist nachfolgend aufgezeigt:


$$(\varphi_n , \varphi_m )  = \frac{1}{L} \int_0^L e^{i\frac{2 \pi k}{L}x} dx \\
$$

$$
(\varphi_n , \varphi_m ) = \frac{1}{L} ( \int_0^L cos(\frac{2 \pi k}{L}x ) dx + \int_0^L i sin(\frac{2 \pi k}{L}x) dx) \\
$$


Das Integral des cos mit den grenzen 0 , L, wird zu 0 und das Integral des sin mit den Grenzen wird 0,L, jeweils zu 1, daher erhalten wir:

 $$(\varphi_n , \varphi_m ) =  \frac{1}{L} (0 -  0 +  1i -1i) =  0 $$

```python
import numpy as np

import matplotlib.pyplot as plt




L = 1.5
D = 0.8

DefBereich = np.linspace(0 ,L ,100)    

dx_4 = L/3
dx_7 =L/6


def fem_laplace_linear_1d_periodic(nb_grid_pts, dx, rhs_x):
#Function to solve the 1D Laplace equation with periodic 
# boundary conditions and an imposed average using a regular grid and linear finite elements. 
#Arguments 
#nb_grid_pts: int # Number of grid points 
   # dx: float #  Length between two adjacent grid points 
   # rhs_x: numpy.ndarray(nb_grid_pts) of floats #       Right-hand-side vector 
 
   # Returns 
   # func_x: numpy.ndarray(nb_grid_pts) of floats 
   #  Solution of the discretized 1D Laplace equation at  each grid point
    
    # K-Matrix definieren
    dig =  2 * np.ones(nb_grid_pts)  #Definition der  Diagonalen 
    ndig = -1 *  np.ones(nb_grid_pts-1)  #Definition der Nebendiagonal. # eins k√ºrzer als hauptgerade

    K =  np.diag(dig) +  np.diag(ndig , 1)  + np.diag(ndig , -1)


    '''√Ñnderung zeilen:'''

    ''' Zeile eins ( periodisch R_Bed)'''
    K[0, : ] =  0
    K[0, -1] = -1
    K[0, 0 ] =  1

    '''last line '''
    K[-1, : ] =  1 
    K[-1, 0 ] = 0.5
    K[-1, -1] = 0.5

    #Vorfactor
    K = 1/dx * K

    return np.linalg.solve(K , rhs_x)

# F√ºr die Rechte Seite:
f_7 = ([0, 0 , 1/D , 0 , -1/D , 0 , (0.3*L) / (dx_7 **2)])
f_4 = ([0, 1/D , -1/D , (0.3*L) / (dx_4 **2)])

# plott
plt.figure()

#Fem L√§esung 7
plt.plot(([0, 0.25, 0.5, 0.75, 1, 1.25, 1.5]), fem_laplace_linear_1d_periodic(7, dx_7, f_7), "r:", label = "N =7")
#Fem L√∂sung 4
plt.plot(([0, 0.5, 1, 1.5]), fem_laplace_linear_1d_periodic(4, dx_4, f_4),"*--", label = "N = 4")




#Analytisch
f_analystisch =  1/D  *  (-np.heaviside(DefBereich - L/3,0) * (DefBereich - L/3) + np.heaviside(DefBereich -2 * L / 3,0) * (DefBereich - 2 * L / 3) +  1/3 * DefBereich + 0.3*D)
plt.plot(DefBereich,f_analystisch,"green", label = "analytisch Loesung")


plt.ylabel("c(x) in m$^{-3}$")
plt.xlabel("Laenge ${L}$ in m")
plt.title(" A 2.3.3")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def FEM_Laplace_Linear_1D(nb_grid_pts, dx, rhs_x, return_system_matrix = False): 
    """ 
    Function to solve the 1D Laplace equation with a 
    Dirichlet boundary condition and a Neumann boundary 
    condition using a regular grid and linear finite 
    elements. 
 
    Arguments 
    --------- 
    nb_grid_pts : int        Number of grid points 
    dx : float Length between two adjacent grid points 
    rhs_x : numpy.ndarray(nb_grid_pts) of floats         Right-hand-side vector 
    return_system_matrix : bool, optional    True if the system matrix should be returned.  (Default: False) 
 
    Returns 
    ------- 
    func_x : numpy.ndarray(nb_grid_pts) of floats 
        Solution of the discretized problem at each grid 
        point 
    system_matrix_xx : numpy.ndarray((nb_grid_pts, nb_grid_pts)) of floats 
        System matrix of the discretized problem. Is only 
        returned if the argument return_system_matrix is 
        True"""
    
    # K-Matrix definieren
    dig =  2 * np.ones(nb_grid_pts)  #Definition der  Diagonalen 
    ndig = -1 *  np.ones(nb_grid_pts - 1)  #Definition der Nebendiagonal

    K =  np.diag(dig) +  np.diag(ndig , 1)  + np.diag(ndig , -1)

    #Zeilen anpassen laut Aufgabe

    #Zeile 1 (Randbedinung links)
    K[0, :] = 0
    K[0, 0] = 1

    #Letze Zeile ( Randbedingung rechts)
    K[-1, :] =  0 
    K[-1,-1] =  1
    K[-1,-2] = -1

    #Vorfactor
    K = 1/dx * K

    #R√ºckgabe
    if return_system_matrix:
        return K

    #loesungsverktor
    return np.linalg.solve(K,rhs_x)




#Parameter 
L = 3
D_constant = 0.8


f_0 = 0.5
c_0_left = 2
c_L_right = 0.7

dx_4  = L / 3
dx_7  = L / 6
dx_10 = L / 9 


Defbereich= np.linspace(0,L,1000)  #100



#konsante Quelleterme
f_4konstant  = ([c_0_left/dx_4,   -f_0 * dx_4/D_constant ,   -f_0 * dx_4/D_constant ,    c_L_right - 1/D_constant  * f_0 * dx_4 / 2])
f_7konstant  = ([c_0_left/dx_7,   -f_0 * dx_7/D_constant ,   -f_0 * dx_7/D_constant ,   -f_0 * dx_7/D_constant ,   -f_0 * dx_7/D_constant ,   -f_0 * dx_7/D_constant ,  c_L_right - 1/D_constant  * f_0 * dx_7/2 ])
f_10konstant = ([c_0_left/dx_10,  -f_0 * dx_10/D_constant ,  -f_0 * dx_10/D_constant ,  -f_0 * dx_10/D_constant ,  -f_0 * dx_10/D_constant ,  -f_0 * dx_10/D_constant , -f_0 * dx_10/D_constant ,  -f_0 * dx_10/D_constant ,  -f_0 * dx_10/D_constant , c_L_right - 1/D_constant  * f_0 * dx_10/2 ])

#Distibution
f_4 = ([c_0_left/dx_4,  1/D_constant ,  -1/D_constant ,   c_L_right])
f_7 = ([c_0_left/dx_7,     0,            1/D_constant ,       0,           -1/D_constant ,  0,  c_L_right])
f_10 = ([c_0_left/dx_10,   0,               0,           1/D_constant ,          0,         0, -1/D_constant ,  0, 0,  c_L_right])






#Define Data

Data1 = list(range(4))      # [0, 1, 2, 3]

Data2= []                   # [0, 0.5, 1, 1.5, 2, 2.5, 3]
for i in range(0, 7):
    value = i * 0.5
    Data2.append(value)   

Data3 = []                  # [0, 0.33, 0.66, 1, 1.33, 1.66, 2, 2.33, 2.66, 3]
for i in range(10):
    value = i / 3
    Data3.append(value)







plt.figure("Quellterm Konstant")

plt.subplot(1,1,1)

#Loesung Analytisch QUellterm:
x = np.linspace(0,L,100)
c1 = c_L_right - (1/D_constant*f_0*L)
c2 = c_0_left
y = 1 / (2*D_constant) * f_0* x**2 + c1 *x + c2
plt.plot(x,y,label = "Loesung analytisch ermittelt")


#Finite-Elemente-Loesung

plt.plot((Data3), FEM_Laplace_Linear_1D(10, dx_10, f_10konstant), "g:",   label = "N = 10")
plt.plot((Data2), FEM_Laplace_Linear_1D(7, dx_7, f_7konstant),    "r+--", label = "N = 7")
plt.plot((Data1), FEM_Laplace_Linear_1D(4, dx_4, f_4konstant),    "b--",  label = "N = 4")


plt.legend()
plt.xlim(0,3)
plt.title("A 2.4.2 konstanter Quellterm")
plt.xlabel("Laenge in m")
plt.ylabel("Konzentration c  in  m$^{-3}$")
#plt.show()



#-----------------------------------------


#delta = '\u0394'
#plot function
plt.figure("delta Distribution")
plt.subplot(1,1,1)

#Finite-Elemente-Loesung
plt.plot((Data3), FEM_Laplace_Linear_1D(10, dx_10, f_10), "+--", label = "N = 10")
plt.plot((Data2), FEM_Laplace_Linear_1D(7,  dx_7,  f_7 ), "b--", label = "N = 7")
plt.plot((Data1), FEM_Laplace_Linear_1D(4,  dx_4,  f_4 ),"red",  label = "N = 4")





plt.legend()
plt.xlim(0,3)
plt.title("A 2.4.2 delta distribution")
plt.xlabel("Laenge in m")
plt.ylabel("Konzentration c  in m$^{-3}$")
plt.show()


```
