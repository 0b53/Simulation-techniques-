#   1.1 Die Bilanzgleichung
In unserem Zweidimensionalen System wird die Teilchenzahl beschrieben durch:
$$N_{ tot} = \int  c ( \vec{r} )  d^2   r$$

Teilchen können nicht erschaffen oder vernichet werden. Daher ist $\dot{N}_{tot} = 0$.

In einem Teilabschnitt des Systems kann sich jedoch die Teilchenzahl ändern, da die Konzentration in dem komplette Gebiet nicht konstant sein muss.
Für diese Änderung gilt daher:
$$\dot{N} = \int_{A} \frac{ \partial c( \vec {r},t)}{\partial t}\vec{r}  d^2  \tag{1}$$

Wie zuvor erwähnt, könnten keine Teilchen vernichtet oder erzeugt werden.Daher bleibt nur, dass du Teilchen über die Kanten ein und austreten.
Die rechte und Linke Seite werden als $\Delta y$ bezeichnet. Die obere und untere Seite werden als $\Delta x$ bezeichnet.

$$
\dot{N} = -j_{rechts} \Delta y - j_{links}\Delta y - j_{oben} \Delta x - j_{unten} \Delta x
\tag{2}
$$

Teilchenstrom Vektroriell betrachtet:
$$
j_{rechts} = j_x(x+\frac{\Delta x}{2}, y)\\
j_{links } = j_x(x-\frac{\Delta x}{2}, y)
$$
$$
j_{oben}   = j_y(x, y + \frac{\Delta y}{2})\\
j_{unten}  = -j_y(x, y - \frac{\Delta y}{2}) \\
$$


Mit Hilfe der Gleichugn zwei und 
$A = \Delta x \cdot \Delta y$ ergibt sich:

$$\dot{N}  =  - \frac{A}{\Delta y} \left(j_y(x,y+\frac{\Delta y}{2}) - j_y(x,y-\frac{\Delta y}{2} )\right) - \frac{A}{\Delta x} \left(j_x(y,x+\frac{\Delta x}{2}) - j_x(y,x-\frac{\Delta x}{2})\right) \\$$
 $$=-\int_{A} \nabla \cdot \vec{j}(\vec{r}) d^2r  \tag{3} $$


Formel (1) und Formel (2) werden gleichgesetzt:
 $$\int_{A} (\frac{\partial c(\vec{r},t)}{\partial t} + \nabla \vec{j}(\vec{r})) d^2 r = 0 $$

Das Argument muss im Integral Null ergeben. Denn es muss für jedes beliebe Flächenelement gelten.

$$
\frac{\partial c(\vec{r},t)}{\partial t} + \nabla \vec{j}(\vec{r}) = 0
\tag{4}
$$


Bei der Annahme, das nun doch Teilchen vernichtet udn erzeugt werden können benötigen wir den Quellterm Q. 
Ist Q größer Null, werden Teilchen produziert. Ist Q kleiner 0, werden Teichen vernichtet. 
Damit folgt für den Instatonären Fall :
$$
\frac{\partial c(\vec{r},t)}{\partial t} +\nabla \vec{j}(\vec{r}) =  Q
$$

Da es im Stationären Fall keine Zeitliche Änderung hinsichtlich der Konzentration gibt, ist die Änderung der Konzentration gelich Null. 
  Daraus folgt:
     $$\nabla \vec{j} ( \vec{r}) = Q $$


#  Aufgabe 1.2: Der Diffusionsstrom
 
Das ein Teilchen sich zum Zeitpunkt $t+ \tau$ an einem Ort x befindet ist durch die Wahrscheinlichkeit P geben:
$$
P(x,t + \tau) = qP (x+h,t)+pP(x-h,t)
$$
$\Delta h$ beschreeibt hier die Sprungweite.
$q$ und $p$ beschreibt die Wahrscheinlichkeiten, dass ein Teilchen von rechts nach links bzw. links nacht rechts springt.

Dadurch, dass q & p jeweils einhalb sind, also gleich groß,folgt das beide Ereignisse gleich Wahrscheinlich sind und somitdie Teilchen während der sprungdauer $\tau$ ihren Ort wechseln.

Stetzt man nun also jeweils für q & p = 1/2 ein, erhält man:
$$
P(x,t + \tau) =  \frac{1}{2} \cdot P (x+h,t)+  \frac{1}{2}\cdot P(x-h,t)
$$
Wenn man nun die Wahrscheinlichkeit P(x,t) auf beiden Seiten der Gleichung  abzieht , also nimmt an dass sich bereits ein Teilchen am Punkt x zum Zeitpunkt t befindet.
Darauf hin auch durch die Sprungdauer $\tau$  dividiert erhält man :

$$
\frac{P(x,t+\tau) - P(x, t)} {\tau} = \frac{P(x+\Delta h,t) - 2P(x,t)+P(x-\Delta h,t)}{2\tau}\\
$$

Abschließend wird die rechte Seite der Gleichung noch mit $\frac{\Delta h^2}{\Delta h^2}$  erweitert.

Reduziert man die Sprungzeit $\tau$ und Schrittweite $\Delta h^2$ erhält man ein kontinuierliches system.
Über eine Taylorreihen Entwicklung erhält man dann das Zweite Ficksche Gesetz.

$$
\frac{\partial c( x,t)}{\partial  t} = D \cdot \frac{\partial ^2 c(x,t)}{\partial x^ 2}\tag{5}
$$


Da wir von einer sehr Hohen Teilchenzahl ausgehen, $N_A \approx 6x 10^{23}$, können wir annehmen, dasss es sich nun nicht mehr um die Besetzungswahrscheinlicheit sondern um eine Konzentration handelt.
Nutzt man nun die Kontinumsglechung ohne Quelleterm (4) und setzt diese mit (5) gleich erhalten wir:

$$
-\nabla  j(x) = D \frac {\partial^ 2 c(x,t)}{ \partial x^2}
$$

Durch Integration erhält man das Erste Ficksche Gesetzt:

$$
j(x) = -D \frac{ \partial c(x ,t)}{ \partial x}\tag{6}
$$

```python
# A 1.1.4

import numpy as np
import matplotlib.pyplot as plt

D = 0.8
L = 3



## FurierReihe
def c_n(n):
    if (n == 0):
        return 0.3       
    return -L/(2 * np.pi**2 * n**2 * D)* 1j * np.sin(2 * np.pi * n/3)
#j = to represent complex number


def FurierReihenKonz(x, n) :
    i = -n 
    c = np.zeros(x.shape)
    #creates an array of zeros with the same shape as the input array x
    while (i <= n) :                
        c= c + np.real(c_n(i)* np.exp(2j * np.pi * i * x/3))
        i += 1
    return c
    
x = np.linspace(0, 3, 600) # Start,Endpunkt u number of Steps
n = 1

y1 = 1/D  *  (-np.heaviside(x - L /3,0) * (x-L/3) + np.heaviside(x - 2*L/3,0)*(x-2*L/3) + 1/3*x + 0.3*D)
y2 = FurierReihenKonz(x, 1)
y3 = FurierReihenKonz(x, 5)
y4 = FurierReihenKonz(x, 10)
y5 = FurierReihenKonz(x, 20)

plt.figure("Blatt1, A1.4")
plt.plot(x,y1, label = "n = Analystisch", color= 'red')
plt.plot(x,y2, label = "n = 1")
plt.plot(x,y3, label = "n = 5")
plt.plot(x,y4, label = "n = 10")
plt.plot(x,y5, label = "n = 20", color = 'yellow')

plt.ylabel("Konzentration in  m$^{-3}$ ")
plt.xlabel("Länge in m ")
plt.legend()
plt.show()
```

#  Aufgabe 1.6: Linearisierte Poisson Boltzmann Gleichung

Die Nerst Planck Gleichung lautet:
$$
\vec{j}_i = -D_i (\nabla c_i + \frac{q_i}{k_bT} c_i \nabla \Phi)
$$

Die Gleichugn für das Chemische Potential lautet:
$$
\mu _i = q_i \Phi + (k_bT) ln(c_i)
$$

Der fordere Teil entspricht der Potentiellen Energie und der hintere Teil der Energie eines idealten Gases.
Setzt man nun die Nerst Planck Gleichung ein, erhält man die Stromdichte:

$$
\vec{j}_i = -\frac{D_i} {k_BT}c_i\nabla \mu_i
$$
  Die Gleichung enspricht dem Stationären Fall wenn der Teilchen Strom gelich nulll ist und somit das Chemische Potentila konstant ist. Somit gilt:
    
 
$$
 c_i  = c_o e^{- \frac{q_i \Phi}{k_b T}}
$$

Setzt man dies nun in deie Poission -Blotzmann Gleichugn ein:
    
$$
\nabla ^2 \Phi = - \frac{1}{\varepsilon} \sum_{n=1}^i q_i c_{i,0}  e^{- \frac{q_i \Phi}{k_B T}}
$$

Druch eine Tayloreihen Entwicklung ergibt sich:
$$
\nabla^2 \Phi = \frac{1}{\epsilon} \cdot \sum_{i = 1}^i q_i c_{i,0}- (1 - \frac{q_i}{k_bT}\Phi)
$$

Das Systema in der Aufgabe ist gegeben durch $N = 2;  ~~~~q_1 = -q_2 = q; ~~~~~c_{1,0} = c_{2,0} = c_{0}$.Damit erhält man :

$$
\nabla ^2 \Phi =   c \cdot \frac{2 q^2 \Phi}{k_b T \varepsilon}
$$

Die Debeilänge ist  $\lambda = \sqrt{\frac{\epsilon k_bT} {2q^2c}}$. Sie ist das elektrische potential welches bei der Entfernung 1/e. Somit erhalten wir:

$$
\nabla ^2\Phi = \frac{1}{\lambda ^2}\Phi
$$

Mit den gegeben Randbedingugnen und dem Exponential Ansatz erhält man:
$$
\Phi = A e(x/\lambda) + B e(-x/\lambda)
$$


$$
\Phi = (-\Phi _0  \frac{\Phi _1 - \Phi _0 e^{\frac{L}{\lambda}}}{e^{- \frac{L}{\lambda}} - e^{\frac{L}{\lambda}}} ) \cdot e^{\frac{x}{\lambda}} + (\frac{\Phi _1 - \Phi _0 e^{\frac{L}{\lambda}}}{e^{-\frac{L}{\lambda}} - e^{\frac{L}{\lambda}}}) \cdot e^{-\frac{x}{\lambda}}
$$



```python
import matplotlib.pyplot as plt
import numpy as np


Debye = 20 ** (-1/2)
L = Debye
T = 300 #Kelvin
q = 1.6 * 10 ** -19
kB = 1.38 * 10 ** -23


phi_0 = -kB * T /q
phi_1 = kB * T / q


B = (phi_1 - phi_0*np.exp( L / Debye)) / (np.exp(-L/Debye) - np.exp(L / Debye))
A = phi_0 - B

x = np.linspace(0, 1 , 300) # Start, Endpunkt, schritte
y = A*np.exp(x) + B * np.exp(-x)

plt.figure( "A 1.6")
plt.plot (x,y)
plt.xlabel(" normalisierte Länge in x/L (relativ)")
plt.ylabel( "normalisiertes elektrisches Potentztial in  $\Phi /  \Phi$")
plt.show()
```
