# A3.1.1 Nichtlinerare Possion Blotzmann Gleichung

Nichtlineare Poisson Blotzmanngleichung:
$$
\nabla^2 \Phi =-  \frac{1}{\epsilon} \sum_\alpha q_\alpha \space c^\infty \space\text{e}^{(\frac{-q_\alpha \Phi}{k_BT})}
$$
Zur Eindimensionalisierung benötigt es ein paar Umstellungen:
$$
\tilde{\Phi} = \Phi q_e / k_BT \\
\tilde{q}_\alpha = q_\alpha / q_e \\
\tilde{x}        = x/\lambda \\
\tilde{y}        = y/\lambda \\
$$

Daraus folgt der die Änderung des Nabla Operators:



$$
\nabla^2 \Phi = (\frac{\partial}{\partial x} \frac{\partial}{\partial y})^2
$$
$$
 = \tilde{\Phi} \tilde{\nabla}^2  \frac{k_BT}{\lambda^2 q_e} 
$$



Damit erhält man auf der Rechten Seite der nichtlineraren Possion Botzmanngleichung durch einsetzten und umformen :



$$
- \frac{1}{\epsilon} \sum_\alpha q_\alpha~\space c^\infty~\text{e}^{(\frac{-q_\alpha \space\Phi}{k_BT})}
$$

$$
-\frac{1}{\epsilon} \sum_\alpha \tilde{q}_\alpha  \space q_e \space c^\infty \space e^{(\frac{-\tilde{q}_\alpha \space q_e \space \tilde{\Phi} \space k_BT}{q_e \space k_BT})} 
$$

$$
-\frac{1}{\epsilon} \sum_\alpha \space   \tilde{q}_\alpha \space q_e \space c^\infty e^{(-\tilde{q}_\alpha  \space \tilde{\Phi})}
$$


Die Gleichsetzung der beiden Gleichungen eerhalten wir :


$$
\tilde{\Phi} \tilde{\nabla}^2  \frac{k_BT}{\lambda^2 q_e}= -\frac{1}{\epsilon} \sum_\alpha \space   \tilde{q}_\alpha \space q_e \space c^\infty e^{(-\tilde{q}_\alpha \tilde{\Phi})}
$$

Die Debyelänge ist definiert als $\lambda = \sqrt{\frac{\epsilon \cdot k_BT} {2 \cdot q_e^2 \cdot c^\infty }}$

dadurch kann man vereinfachen zu 

$$
\tilde{\nabla^2} \tilde{\Phi} = \frac{-1}{2} \sum_{\alpha} \tilde{q}_{\alpha} e^{(- \tilde{q}_{\alpha}} \tilde{\Phi})
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def laplace_element_matrix(dx,dy):
    """
    Construct the element matrix of the partially integrated Laplace operator with
    homogeneous Neumann boundary conditions.
    The rows and columns of the element matrix correspond to the nodes (corners of
    the triangle) ordered in geometric positive order, starting with the node with
    the 90° angle.


    Parameters:
   
    dx : float         Grid spacing in x direction
    dy : float         Grid spacing in y direction

    Returns:
      element_matrix_ll : numpy.ndarray        Array of shape (3,3) containing the element matrix
    """
    
    element_matrix = np.zeros((3,3))                                               #Erzeugen der empty  Matrix
    element_matrix[0,: ]  = [(dx/dy + dy / dx)/2, -dy / (dx*2),    -dx / (dy*2)]   #Erste Zeile
    element_matrix[1,: ]  = [-dy/(dx*2),           dy / (dx*2),    0]              #Zweite Zeile
    element_matrix[2,: ]  = [-dx/(dy*2),              0,        dx / (dy*2)]       #Dritte Zeile
    return element_matrix

laplace_element_matrix(2,3)  #Example
```

```python
def node_coordinates(g,  nb_nodes):
    """
    Convert the global linear consecutive index of a node to its Cartesian coordinates.
    
    Parameters:
    g : int        global linear index of the node
    nb_nodes : tuple of ints        Number of nodes in the Cartesian directions --> in der Form (Nx, Ny) (Im Bild: (4, 4))
        
    Returns:  
    x : int        x-coordinate (integer) of the node
    y : int        y-coordinate (integer) of the node
    """
    Nx,  Ny = nb_nodes  #beide nehmen wert von nb_ntes an
    x =   g % Nx          
    y =   g // Nx
    return x, y
```

```python
def node_index(a,  b,  nb_nodes):
    """
    Turn node coordinates (x, y) into their global node index g.

    Parameters:
    x : int        x-coordinate (integer) of the node
    y : int        y-coordinate (integer) of the node
    nb_nodes : tuple of ints        Number of nodes in the Cartesian directions (Nx, Ny)

    Returns
    g : int        Global node index
    """
    Nx, Ny =  nb_nodes 
    return  a + Nx * b
```

```python
import numpy as np
import matplotlib.pyplot as plt

def make_grid(nb_nodes):
    """
    Make an array that contains all elements of the grid. The
    elements are described by the global node indices of
    their corners. The order of the corners is in order of
    the local node index.

    They are sorted in geometric positive order and the first
    is the node with the right angle corner at the bottom
    
    . Elements within the same box are consecutive.

    This is the first element per box:

        2
        | \
        |  \ 
    dy  |   \
        |    \
        0 --- 1
    
          dx
          
    This is the second element per box:
     
           dx
         1 ---0
          \   | 
           \  |  dy 
            \ |  
             \|
              2

    Parameters
    ----------
    nb_nodes : tuple of ints
        Number of nodes in the Cartesian directions

    Returns
    -------
    triangles_el : numpy.ndarray
        Array containing the global node indices of the
        element corners. The first index (suffix _e)
        identifies the element number and the second index
        (suffix _l) the local node index of that element.
    """

    Nx,  Ny  = nb_nodes 

    # Kapitel 11 skript
    # These are the node position on a subsection of the grid 
    # that excludes the rightmost and topmost nodes. The 
    # suffix _G indicates this subgrid. 
    y_G, x_G = np.mgrid[:Ny-1, :Nx-1] 
    x_G.shape = (-1,) 
    y_G.shape = (-1,) 
 
    # List of triangles.
    lower_triangles = np.vstack( 
        (node_index(x_G, y_G, nb_nodes), 
         node_index(x_G+1, y_G, nb_nodes), 
         node_index(x_G, y_G+1, nb_nodes))) 
    upper_triangles = np.vstack( 
        (node_index(x_G+1, y_G+1, nb_nodes), 
         node_index(x_G, y_G+1, nb_nodes), 
         node_index(x_G+1, y_G, nb_nodes))) 
    # Suffix _e indicates global element index.
    return np.vstack( 
        (lower_triangles, upper_triangles)).T.reshape(-1, 3) 
```

```python
def show_mesh(nb_nodes, physical_sizes):
    """
    Show a simple representation of the grid, showing the order of
    the linear indices for nodes and elements.
    
    Parameters:
    
    nb_nodes : tuple of ints
        Number of nodes along x and y direction
        
    physical_sizes : tuple of floats
        Length of the simulation domain along x and y direction
    """

    def nb_elements(nb_nodes):
        nx, ny = nb_nodes
        return (nx - 1) * (ny - 1) * 2

    sx, sy = physical_sizes
    nx, ny = nb_nodes
    dx = sx / (nx-1)
    dy = sy / (ny-1)
    assert np.prod(nb_nodes) < 40, "grid to big to be reasonably plotted"

    
    fig, ax = plt.subplots()
    triangles_el = make_grid(nb_nodes)
    for k in range(nb_elements(nb_nodes)):
        nodes_n = triangles_el[k]
        mean_x = 0
        mean_y = 0
        for n in nodes_n:
            ix, iy = node_coordinates(n, nb_nodes)
            mean_x = mean_x + ix
            mean_y = mean_y + iy
            ax.text(ix * dx, iy * dy, f"n{n}")
        ax.text(mean_x / 3 * dx, mean_y / 3 * dy, f"e{k}", color="red")

        
    ax.set_xlim(-dx, nx * dx)
    ax.set_ylim(-dy, ny * dy)

    ax.set_xlabel("x")
    ax.set_ylabel("y")
```

```python
def plot_colormap(field_g, nb_nodes, physical_sizes,
                  show_mesh=False, levels=11):
    """
    Show a scalar field on a triangulated structured grid using
    a colormap representation. The field will be linearly interpolated
    between the node values.
    
    Parameters
    ----------
    field_g : array_like
        Values of the field in serial form (of length corresponding
        to number of nodes)
    nb_nodes : tuple of ints
        Number of nodes (Nx, Ny) in the Cartesian directions
    physical_sizes : tuple of floats
        Length of the simulation domain along x and y direction
    show_mesh : bool, optional
        If True, shows nodes and element boundaries in black
        (Default: False)
    levels : int, optional
        Number of levels in the colormap
        (Default: 11)
    
    Returns
    -------
    figure : matplotlib.Figure
        Figure object
    axes : matplotlib.Axes
        Axes object
    colorbar : matplotlib.Colorbar
        Colorbar object
    """
    import matplotlib.tri as mtri

    Nx, Ny = nb_nodes
    sx, sy = physical_sizes

    dx = sx / (Nx - 1)
    dy = sy / (Ny - 1)

    fig, ax = plt.subplots()

    ax.set_aspect(1)

    x = np.array([node_coordinates(i, nb_nodes)[0] for i in range(np.prod(nb_nodes))])
    y = np.array([node_coordinates(i, nb_nodes)[1] for i in range(np.prod(nb_nodes))])
    
    x = x * dx
    y = y * dy

    triangles_el = make_grid(nb_nodes)
    triang = mtri.Triangulation(x, y, triangles_el)

    cb = plt.colorbar(ax.tricontourf(triang, field_g, levels=levels))
    if show_mesh:
        ax.triplot(triang, 'ko-')

    ax.set_xlabel(r"x ($\lambda$)")
    ax.set_ylabel(r"y ($\lambda$)")

    return fig, ax, cb
```

```python
def interpolate(x_i, y_i, field_g, nb_nodes, physical_sizes):
    """
    Performs a linear finite element interpolation of the field in x, y 
    
    Parameters: 
    -----------
    x_i, y_i : array_like
        Positions where to interpolate the field
    field_g : array_like
        Values of the field in serial form (of length corresponding
        to number of nodes)
    nb_nodes : tuple of ints
        Number of nodes (Nx, Ny) in the Cartesian directions
    physical_sizes : tuple of floats
        Length of the simulation domain along x and y direction

    Returns: 
    --------
    interpolated_field_i : numpy.ndarray
        Values interpolated at position x_i, y_i
    
    """ 
    
    import matplotlib.tri as mtri

    Nx, Ny = nb_nodes
    sx, sy = physical_sizes

    dx = sx / (Nx - 1)
    dy = sy / (Ny - 1)

    _x = np.array([node_coordinates(i, nb_nodes)[0]
                   for i in range(np.prod(nb_nodes))])
    _y = np.array([node_coordinates(i, nb_nodes)[1]
                   for i in range(np.prod(nb_nodes))])

    _x = _x * dx
    _y = _y * dy

    triangles_el = make_grid(nb_nodes)
    triang = mtri.Triangulation(_x, _y, triangles_el)

    interpolated_field = mtri.LinearTriInterpolator(triang, field_g)

    return interpolated_field(x_i, y_i)
```

```python
#A 3.2.7
#Kapitel 11 skript
def add_element_matrix(system_matrix_gg, element_matrix_ll, 
                       global_node_indices): 
    
    
    """ 
    Add element matrix to global system matrix. 
 
    Parameters 
    ---------- 
    system_matrix_gg : array_like 
        N x N system matrix where N is the number of global 
        nodes. This matrix will be modified by this function. 
    element_matrix_ll : array_like 
        n x n element matrix where n is the number of local 
        nodes 
    global_node_indices : list of int 
        List of length n that contains the global node 
        indices for the local node index that corresponds to 
        the list position. 
    """

    
    assert element_matrix_ll.shape == (len(global_node_indices), len(global_node_indices)) 
    for i in range(len(global_node_indices)): 
        for j in range(len(global_node_indices)): 
            system_matrix_gg[global_node_indices[i], 
                             global_node_indices[j]] += element_matrix_ll[i, j]






```

```python
#Kapitel 11 skript
def assemble_system_matrix(element_matrix_ll, nb_nodes): 
    
    """ 
    Assemble system matrix from the element matrix 
 
    Parameters 
    ---------- 
    element_matrix_ll : array_like 
        3 x 3 element matrix 
    nb_nodes : tuple of ints 
        Number of nodes in the Cartesian directions 
 
    Returns 
    ------- 
    system_matrix_gg : numpy.ndarray 
        System matrix 
    """ 
 

    Nx, Ny = nb_nodes 
    Mx, My = Nx-1, Ny-1 # number of boxes 
 
    # System matrix 
    system_matrix_gg = np.zeros([Nx*Ny, Nx*Ny]) 
 
    # Construct system matrix 
    for l in range(Mx): 
        for m in range(My): 
            # Element (0) 
            n0 = node_index(l, m, nb_nodes) 
            n1 = node_index(l+1, m, nb_nodes) 
            n2 = node_index(l, m+1, nb_nodes) 
            add_element_matrix(system_matrix_gg, 
                               element_matrix_ll, 
                               [n0, n1, n2]) 
 
            # Element (1) 
            n0 = node_index(l+1, m+1, nb_nodes) 
            n1 = node_index(l, m+1, nb_nodes) 
            n2 = node_index(l+1, m, nb_nodes) 
            add_element_matrix(system_matrix_gg, 
                               element_matrix_ll, 
                               [n0, n1, n2]) 
 
    return system_matrix_gg

```

```python

def poisson_boltzmann_bulk_matrix(nb_nodes, physical_sizes): 
    
    """ 
    Assemble system matrix 
 
    Parameters 
    ---------- 
    
    nb_nodes : tuple of ints 
        Number of nodes in the Cartesian directions 
    
    physical_sizes : tuple of floats
        Length of the simulation domain along x and y direction
 
    Returns 
    ------- 
    system_matrix_gg : numpy.ndarray 
        System matrix 
    """ 
    
    Nx, Ny = nb_nodes 
    Mx, My = Nx - 1, Ny - 1 # number of boxes 
 
    # System matrix, index g indicates global node
    system_matrix_gg = np.zeros([Nx * Ny, Nx * Ny])
    
    # determine parameters of simulation areas
    length_x, length_y = physical_sizes
    dx = length_x / Mx
    dy = length_y / My
    
    # Consturct local element matrix 
    # --> sum is needed because of 3.2.2 (there is the sum of mass_element_matrix and laplace_element_matrix)
    element_matrix = mass_element_matrix(dx,dy) + laplace_element_matrix(dx,dy)
    
    # Construct system matrix    
    system_matrix_gg = assemble_system_matrix(element_matrix, nb_nodes)
  
    return system_matrix_gg
```

```python
#Herleitung zufinden in Übugsblatt 1 A6.



def poisson_boltzmann_linear_1d_ana(x,  potential_bottom,  potential_top, plate_separation):
    
    """
    Parameters
    ----------
    x :array_like
        Positions where to evaluate the analytical solution.
        In unis of the debye length 
        debye^2 = k_bT*permittivity / (q**2 * concentration_bath)   #Wo ist die 2 im Nenner??!!
    potental_bottom : float 
        Value of the electric potential on the plate at y = 0
        In units of the thermal_voltage = kb*T / q
    potental_top : float
        Value of the electric potential on the plate at y = plate_separation
    plate_separation: float
        distance between the bottom and top plate in units of the debye length
    Returns
    ------- 
    potential : np.ndarray 
    Analytical solution of the 1D linearized Poisson-Boltzman equation 
    """

    K_1 = (potential_top  -  potential_bottom  *  np.exp(-plate_separation)) / (np.exp(plate_separation) - np.exp(  -plate_separation))
    K_2 = potential_bottom  -  K_1
    K_11 = K_1 * np.exp(1*x)
    K_22=  K_2 * np.exp(-1*x)    
    K_ges = K_11 + K_22
   
    
    return K_ges
```

```python

def two_electrode_bc(system_matrix_gg, rhs_g, nb_nodes, potential_bottom, potential_top):
    """
    Modifies the global system matrix and the right hand side according to following boundary conditions:
    bottom and top faces: Dirichlet
    left and right faces: Neumann homogeneous
    
    Parameters
    ----------
    system_matrix_gg : ndarray of shape (Nx * Ny, Nx * Ny)
        System matrix with natural boundary conditions
    rhs_g: ndarray of shape (Nx * Ny)
        Right hand side with natural boundary conditions
    nb_nodes : tuple of ints
        Number of nodes in the Cartesian directions (Nx, Ny)
    potential_bottom : float
        Value of the electric potential on the bottom electrode (y = 0)
    potential_top : float
        Value of the electric potential on the top electrode (y = plate_separation)
     
    """
    Nx, Ny = nb_nodes
    # Dirichlet boundary conditions for top plate 
    for i in range(Nx): 
        n        = node_index(i, Ny - 1, nb_nodes) 
        mat_g    = np.zeros(Nx * Ny) 
        mat_g[n] = 1 
        system_matrix_gg[n] = mat_g 
        rhs_g[n] = potential_top
 
    # Dirichlet boundary conditions for bottom plate 
    for i in range(Nx): 
        n        = node_index(i, 0, nb_nodes) 
        mat_g    = np.zeros(Nx * Ny) 
        mat_g[n] = 1 
        system_matrix_gg[n] = mat_g 
        rhs_g[n] = potential_bottom
        
        
```

```python
#kapitel 11 capicatorbc , nur rechte Seite des Kondensators

def capacitor_bc(system_matrix_gg, rhs_g, 
                 top_left, top_right, top_potential, 
                 bottom_left, bottom_right, bottom_potential, 
                 nb_nodes): 
    """ 
    Set boundary conditions for the parallel plate capacitor. 
 
    Parameters 
    ---------- 
    system_matrix_gg : numpy.ndarray 
        System matrix. The system matrix is modified by a call 
        to this function 
    rhs_g : numpy.ndarray 
        Right-hand side vector. The right-hand side vector is 
        modified by a call to this function. 
    top_left : int 
        Leftmost node of the top electrode 
    top_right : int 
        Rightmost node of the top electrode 
    top_potential : float 
        Electrostatic potential of the top electrode 
    bottom_left : int 
        Leftmost node of the bottom electrode 
    bottom_right : int 
        Rightmost node of the bottom electrode 
    bottom_potential : float 
        Electrostatic potential of the bottom electrode 
    nb_nodes : tuple of ints 
        Number of nodes in the Cartesian directions 
    """ 
    Nx, Ny = nb_nodes 
    # Dirichlet boundary conditions for top plate 
    for i in range(top_left, top_right+1): 
        n = node_index(i, Ny-1, nb_nodes) 
        mat_g = np.zeros(Nx*Ny) 
        mat_g[n] = 1 
        system_matrix_gg[n] = mat_g 
        rhs_g[n] = top_potential 
 
    # Dirichlet boundary conditions for bottom plate 
    for i in range(bottom_left, bottom_right+1): 
        n = node_index(i, 0, nb_nodes) 
        mat_g = np.zeros(Nx*Ny) 
        mat_g[n] = 1 
        system_matrix_gg[n] = mat_g 
        rhs_g[n] = bottom_potential
        
```
